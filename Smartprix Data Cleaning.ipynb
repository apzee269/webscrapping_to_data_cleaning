While reviewing the data manually, several issues have been identified across various columns.

Let's categorize each column issues into two types: Quality Issues and Tidiness Issues.

1) <b>Quality Issues </b> refers to data points that is accurate in terms of information but is in the wrong format or stored in the wrong column.
2) <b>Tidiness Issues</b>, on the other hand, refers to data points that contains missing, inaccurate, or duplicate values.

Quality Issues can be further categorize into four aspects:

1) <b>Completeness</b>: Data that has missing values.
2) <b>Validity</b>: Data that is not valid, such as having negative values in columns for price or rating.
3) <b>Accuracy</b>: Data that is valid but contains inaccurate information, such as a feature phone being classified as a smartphone.
4) <b>Consistency</b>: Data that is written differently across different entries, such as Samsung being written as "Samsung", "SAMUNG", or "samsung".

The camera_extractor function takes a text string as an input and returns a string that represents the number of cameras based on the text. If the text contains the word 'Quad', the function returns '4', if the text contains the word 'Triple', the function returns '3', if the text contains the word 'Dual', the function returns '2', if the text contains the word 'Missing', the function returns '0', and if none of the above conditions are met, the function returns '1'.

```python
def camera_extractor(text):
    if 'Quad' in text:
        return '4'
    elif 'Triple' in text:
        return '3'
    elif 'Dual' in text:
        return '2'
    elif 'Missing' in text:
        return '0'
    else:
        return '1'
```

```python
camera_num = df['camera'].apply(camera_extractor)
```

```python
num_front_cameras = df['camera'].str.strip().str.split('&').str.get(1).str.strip().fillna('Missing').apply(camera_extractor)
```

```python
df.insert(23,'num_back_camera',camera_num)
df.insert(24,'num_front_camera',num_front_cameras)
```

One-Hot Encoding is a technique that converts categorical data into a binary format that can be used in machine learning. It's used when working with categorical data that can't be easily represented numerically, and is often used with decision trees, random forests, and neural networks.

Removing "model" column as it will not add any details into our analysis

This code creates a new DataFrame by performing one-hot encoding on a set of categorical columns in a given DataFrame (df).

The list of columns to encode is defined in the cols variable. The for loop iterates through each column and applies the pd.get_dummies() function to create a new DataFrame with binary columns for each unique value in the original categorical column.

The resulting binary columns are concatenated horizontally using pd.concat() with the argument axis=1 to combine them with the previous set of encoded columns. The resulting DataFrame a contains binary columns for each unique value in the categorical columns in the original DataFrame, which can be used as input to machine learning models that require numerical data.

```python
cols = ['brand','sim_type','5g','volte','nfc','ir_blaster','processor_brand','display_type','display_category','os']
a = pd.DataFrame()
for i in cols:
    a = pd.concat([a,pd.get_dummies(df[i])],axis = 1)
                   
```

```python
df = pd.concat([df,a],axis = 1)
```
